# Beginning Visualizations

```python
df.info()
```

```python
df.value_counts()
```

```python
#This cannot be a scatter, it's the wrong variable types.... I'm leaving the code to see it tho
zurichClassVsArea = pd.DataFrame({'x':df["zurich_class"], 'y':df["area"]})

plt.scatter(zurichClassVsArea['x'], zurichClassVsArea['y'])

plt.xlabel("Zurich Class")
plt.ylabel("Area")
plt.show()
```

```python
#categorical and numerical... needs to be a bar graph

zc_v_area_counts = df.groupby("zurich_class")["area"].count()
plt.bar(zc_v_area_counts.index, zc_v_area_counts.values)

plt.xlabel("Zurich Class")
plt.ylabel("Area")
plt.show()
```

```python
zurichClassVsCClassFlares = pd.DataFrame({'x':df["zurich_class"], 'y':df["c_class_flares"]})

c_class_counts = zurichClassVsCClassFlares["y"].value_counts()

plt.bar(c_class_counts.index, c_class_counts.values)
plt.xlabel("Zurich Class")
plt.ylabel("C Class Flares")
plt.show()
```

```python
#Z Class v C class flares scatter...
zurichClassVsCClassFlares = pd.DataFrame({'x':df["zurich_class"], 'y':df["c_class_flares"]})

plt.scatter(zurichClassVsCClassFlares['x'], zurichClassVsCClassFlares['y'])
plt.xlabel("Zurich Class")
plt.ylabel("C Class Flares")
plt.show()
```

```python
zurichClassVsLargestSpotSize = pd.DataFrame({'x':df["zurich_class"], 'y':df["largest_spot_size"]})

plt.scatter(zurichClassVsLargestSpotSize['x'], zurichClassVsLargestSpotSize['y'])
plt.xlabel("Zurich Class")
plt.ylabel("Largest Spot Size")
plt.show()
```

```python
ActivityVsEvolution = pd.DataFrame({'x':df["activity"], 'y':df["evolution"]})

plt.scatter(ActivityVsEvolution['x'], ActivityVsEvolution['y'])
plt.xlabel("Activity")
plt.ylabel("Evolution")
plt.show()
```

```python
zurichClassVsSpotDistribution = pd.DataFrame({'x':df["zurich_class"], 'y':df["spot_distribution"]})

plt.scatter(zurichClassVsSpotDistribution['x'], zurichClassVsSpotDistribution['y'])
plt.xlabel("Zurich Class")
plt.ylabel("Spot Distributions")
plt.show()
```
